```python
import folium
```

```python
import pandas as pd

df = pd.read_csv('/content/Cura por Municipio-3.csv', encoding='latin-1')

print(df.to_string())
```

```python
geo_json_url = '/content/regioes.geojson'
```

```python
mapa_cura_tb = folium.Map([-29.478182, -52.927251], tiles = 'cartodbpositron', zoom_start = 7)
```

```python
mapa_cura_tb
```

```python
#Mapa Cloroplético

folium.Choropleth(geo_data = geo_json_url, data = df,
                  columns = ["REG_NOME", "Valor2019"],
                  key_on = 'feature.properties.REG_NOME',
                  fill_color= "GnBu",
                  #fill_opacity= 0.9,
                  #line_opacity= 0.5,
                  #legend_name= "Porcentagem Cura 2019"
                  nan_fill_color= "red"
                  ).add_to(mapa_cura_tb)



#mapa_cura_tb.save("Cura 2019.jpeg")
```
